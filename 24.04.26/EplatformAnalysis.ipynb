```python
from pyspark.sql.window import Window
from pyspark.sql.functions import col, rank, dense_rank, row_number, sum
```

```python
window_spec = Window.orderBy(col("price").desc())

products.withColumn("rank", rank().over(window_spec)).show()
```

```python
window_spec = Window.partitionBy("category").orderBy(col("price").desc())

products.withColumn("rank", rank().over(window_spec)).show()
```

```python
window_spec = Window.orderBy("signup_date")

customers.withColumn("row_num", row_number().over(window_spec)).show()
```

```python
full_df_rev = full_df.withColumn("revenue", col("quantity") * col("price"))

window_spec = Window.orderBy(col("revenue").desc())

full_df_rev.withColumn("rank", rank().over(window_spec)).show()
```

```python
window_spec = Window.orderBy(col("revenue").desc())

full_df_rev.withColumn("dense_rank", dense_rank().over(window_spec)).show()
```

```python
from pyspark.sql.functions import sum as _sum

customer_rev = full_df.withColumn("revenue", col("quantity") * col("price")) \
                      .groupBy("city", "name") \
                      .agg(_sum("revenue").alias("total_revenue"))

window_spec = Window.partitionBy("city").orderBy(col("total_revenue").desc())

customer_rev.withColumn("rank", rank().over(window_spec)) \
            .filter(col("rank") == 1) \
            .show()
```

```python
product_rev = full_df.withColumn("revenue", col("quantity") * col("price")) \
                     .groupBy("category", "product_name") \
                     .agg(_sum("revenue").alias("total_revenue"))

window_spec = Window.partitionBy("category").orderBy(col("total_revenue").desc())

product_rev.withColumn("rank", rank().over(window_spec)) \
           .filter(col("rank") == 1) \
           .show()
```

```python
window_spec = Window.orderBy("order_date")

full_df.withColumn("revenue", col("quantity") * col("price")) \
       .withColumn("running_total", sum("revenue").over(window_spec)) \
       .show()
```

```python
window_spec = Window.orderBy(col("amount").desc())

payments.withColumn("rank", rank().over(window_spec)).show()
```

```python
spark.sql("""
SELECT c.name, o.order_id, o.status
FROM customers c
JOIN orders o
ON c.customer_id = o.customer_id
""").show()
```

```python
spark.sql("""
SELECT c.name, COUNT(o.order_id) AS total_orders
FROM customers c
JOIN orders o
ON c.customer_id = o.customer_id
GROUP BY c.name
""").show()
```

```python
spark.sql("""
SELECT p.product_name, oi.quantity
FROM order_items oi
JOIN products p
ON oi.product_id = p.product_id
""").show()
```

```python
spark.sql("""
SELECT oi.order_id,
       SUM(oi.quantity * p.price) AS revenue
FROM order_items oi
JOIN products p
ON oi.product_id = p.product_id
GROUP BY oi.order_id
""").show()
```

```python
spark.sql("""
SELECT c.name,
       SUM(oi.quantity * p.price) AS revenue
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_items oi ON o.order_id = oi.order_id
JOIN products p ON oi.product_id = p.product_id
GROUP BY c.name
ORDER BY revenue DESC
LIMIT 3
""").show()
```

```python
spark.sql("""
SELECT p.product_name,
       SUM(oi.quantity * p.price) AS revenue
FROM order_items oi
JOIN products p
ON oi.product_id = p.product_id
GROUP BY p.product_name
ORDER BY revenue DESC
LIMIT 1
""").show()
```
