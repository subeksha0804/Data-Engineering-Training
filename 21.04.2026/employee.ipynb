```python
!pip install pyspark
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("employee").getOrCreate()
```

```python
data=[
    ("sube",50000,"dataeng"),
    ("sushan",60000,"finance"),
    ("suba",50000,"dataeng"),
    ("aravind",60000,"marketing"),
    ("vetha",60000,"hr")
]
colums=["name","salary","dept"]
df = spark.createDataFrame(data,colums)
df.show()
```

```python
df.select("name","salary").show()
```

```python
df.filter(df.salary>55000).show()
```

```python
df.groupby("dept").count().show()
```

```python
df.orderBy("salary").show()
```
