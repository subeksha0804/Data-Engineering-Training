```python
sales_df = spark.read.csv("sales.csv",header=True,inferSchema=True)
sales_df.show()
```

```python
from pyspark.sql.functions import col

sales_df = sales_df.withColumn("revenue", col("quantity") * col("price"))
sales_df.show()
```

```python
sales_df.select(sum("revenue")).show()
```

```python
total_sale =sales_df.groupBy("product").sum("quantity")
total_sale.show()
```

```python


sales_df.groupBy("product") \
        .sum("quantity") \
        .orderBy(desc("sum(quantity)")) \
        .show(1)
```

```python
sales_df.groupBy("emp_id").sum("revenue").orderBy(desc("sum(revenue)")).show(1)
```

```python
sales_df.select(avg("revenue")).show()
```

```python
sales_df.groupBy("product") \
        .sum("revenue") \
        .filter("sum(revenue) > 100000") \
        .show()
```

```python
# Read the JSON file with multiline option
df = spark.read.option("multiline", "true").json("orders.json")

# Import explode to flatten the orders array
from pyspark.sql.functions import explode

# Explode the orders array and select individual fields
df = df.select(explode("orders").alias("order")).select("order.*")

# Display the DataFrame
df.show()
```

```python
df.select(sum("amount")).show()
```

```python
df.groupBy("customer").sum("amount").show()
```

```python
df.groupBy("customer") \
  .sum("amount") \
  .orderBy(desc("sum(amount)")) \
  .show()
```

```python
df.groupBy("product").sum("amount").show()
```

```python
df.filter("city='Hyderabad'").show()
```
